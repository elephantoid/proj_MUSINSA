# Adding CI

```python
import  pymysql
db= pymysql.connect(host='localhost', user='root', password='root', db='mall')
curs =db.cursor(pymysql.cursors.DictCursor)
query= "select seq, ssn from users"
curs.execute(query)
rows =curs.fetchall()
for row in rows:
    print(row["seq"], row["ssn"])
db.close()
```

```python
import hmac
import hashlib
import binascii
ssn = rows[0]['ssn']
RNstr = ssn[0:6] + ssn[7:16]
RNbyte = RNstr.encode('utf-8')
RNbyte2 = binascii.b2a_hex(RNbyte)
Padding ='%0102x' % int(0)
temp1 = RNbyte2.decode('utf-8') + Padding
len(temp1)
```

```python
import random

SA ='%02x' %random.getrandbits(512)
Sk ='%02x' %random.getrandbits(512)

print(SA)
print(Sk)
```

```python
temp2 = int(temp1, 16) ^int(SA, 16)
temp3 = ('%02x' % temp2).encode('utf-8')
CI = hmac.new(Sk.encode('utf-8'), temp3, hashlib.sha512)
CI.hexdigest()
```
