```python
import random

credit_card_type=[
    'SSG.COM카드','이마트e현대카드', '현대카드', '이마트KB국민카드', 'KB국민카드',
    '이마트삼성카드', '신세계삼성카드', '트레이더스삼성카드', '삼성카드', '이마트신한카드',
    '신세계신한카드', '신한카드','비씨카드','신세계하나체크카드', '하나카드', '롯데카드', 'NH카드',
    '카카오뱅크카드', '신세계씨티카드', '씨티카드', '이마트우리체크카드', '우리카드', 'IBK기업은행카드', '이마트SC카드',
    '신세계SC카드', 'SC은행카드', 'SSGPAY카드', '광주카드', '수협카드', '제주카드', '전북카드'
]
for i in range(10):
    print(random.choice(credit_card_type))
```

```python
from faker import Faker
import pymysql
fake =Faker('ko_KR')
Faker.seed(0);

db = pymysql.connect(host='localhost', user='root', password='root', db='mall')
curs = db.cursor()
seq = 0
for users_seq in range(1000):
    for _ in range(random.randint(1,3)):
        number= fake.credit_card_number()
        expiration_date = fake.credit_card_expire()
        cvc = fake.credit_card_security_code()
        name = fake.name()
        type1 = random.choice(credit_card_type)
        query = "INSERT INTO credit_cards(users_seq, number, expiration_date, cvc, name, type)"\
        +"VALUES("+str(users_seq+1)+",'"+str(number)+"','"\
        +expiration_date +"',"+str(cvc)+", '"+name+"', '"+type1+"');"
        print(query)
        seq+=1
        curs.execute(query)
db.commit()
db.close
print(seq)
```
