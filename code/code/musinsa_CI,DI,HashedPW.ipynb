```python
import hmac
import hashlib
import binascii
import random
SA ='%02x' %random.getrandbits(512)
Sk ='%02x' %random.getrandbits(512)
CP = '%02x' % random.getrandbits(96)
CIs =[]
DIs =[]
hashedpws = []
for i in range(len(rows)):
    ssn = rows[i]['ssn']
    RNstr = ssn[0:6] + ssn[7:16]
    RNbyte = RNstr.encode('utf-8')
    RNbyte2 = binascii.b2a_hex(RNbyte)
    #CI
    Padding ='%0102x' % int(0)
    temp1 = RNbyte2.decode('utf-8') + Padding
    temp2 = int(temp1, 16) ^int(SA, 16)
    temp3 = ('%02x' % temp2).encode('utf-8')
    CI = hmac.new(Sk.encode('utf-8'), temp3, hashlib.sha512)
    print(i+1,"CI [", CI.hexdigest(), "]")
    CIs.append(CI.hexdigest())
    #DI
    Padding2 = "%087x" % int(0)
    temp4 = str(RNbyte2.decode('utf-8'))+CP+Padding2
    DI= hashlib.sha512(temp1.encode())
    DI.hexdigest()
    print(i+1,"DI [", DI.hexdigest(), "]")
    DIs.append(CI.hexdigest())
    #hashed Pw
    salt ='%02x' % random.getrandbits(8)
    password = rows[i]['password']
    temp5= salt+password
    hashed = hashlib.sha256(temp5.encode('utf-8')).hexdigest()
    hashedpw = salt+hashed
    print(i+1, "hashedpw [", hashedpw,"]")
    hashedpws.append(hashedpw)
```

```python
import  pymysql
db= pymysql.connect(host='localhost', user='root', password='root', db='mall')
curs =db.cursor(pymysql.cursors.DictCursor)
for i in range(len(rows)):
    query ="UPDATE users SET CI=unhex('"+ CIs[i]+"') WHERE seq="+str(i+1)+";"
    curs.execute(query)
    query2 ="UPDATE users SET DI=unhex('"+ DIs[i]+"') WHERE seq="+str(i+1)+";"
    curs.execute(query2)
    query3 ="UPDATE users SET hashedpw=unhex('"+ hashedpws[i]+"') WHERE seq="+str(i+1)+";"
    curs.execute(query3)
db.commit()
db.close()
```
